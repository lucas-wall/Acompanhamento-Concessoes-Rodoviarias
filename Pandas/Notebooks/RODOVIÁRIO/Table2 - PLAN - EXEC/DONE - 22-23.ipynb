```python
import pandas as pd
import openpyxl
import numpy as np
```

```python
file_path1 = 'NEW-SCRIPT - LIMPO\RAW_PLAN_EXEC ATÉ ANO ANTERIOR (22).xlsx'
file_path2 = 'NEW-SCRIPT - LIMPO\RAW_META(2023).xlsx'
file_path3 = 'NEW-SCRIPT - LIMPO\RAW_META(2024).xlsx'
```

```python
# Carregar as planilhas


df_plan_exec_ano_ant_raw = pd.read_excel(file_path1)
df_meta_23_raw = pd.read_excel(file_path2)
df_meta_24_raw = pd.read_excel(file_path3)
```

<h1><b><i>Tabela - EXECUTADO - META 23</i></b></h1>

```python
df_concatenado = pd.concat([df_plan_exec_ano_ant_raw, df_meta_23_raw], ignore_index=True)
```

```python
df_concatenado
```

```python
# Drop na coluna 'Região' que não é usada.
df_concatenado.drop(columns=['ID-ÚNICO'], inplace=True)
```

```python
df_concatenado.index = range(1, len(df_concatenado) + 1)
df_concatenado.index.name = 'ID-ÚNICO'
df_concatenado = df_concatenado.reset_index()
```

```python
df_concatenado = df_concatenado.sort_values(by='ID-ÚNICO', ascending=True)
```

```python
new_column_names = ['PAVIMENTAÇÃO - Descrição',
 'PAVIMENTAÇÃO - km(i)', 
 'PAVIMENTAÇÃO - km(f)', 
 'PAVIMENTAÇÃO - Ext.(km)', 
 'PAVIMENTAÇÃO - PERCENTUAL (%) PLAN', 
 'PAVIMENTAÇÃO - (km)% PLAN', 
 'PAVIMENTAÇÃO - FINANCEIRO PLAN (R$)', 
 'PAVIMENTAÇÃO - PERCENTUAL (%) EXEC', 
 'PAVIMENTAÇÃO - (km)% EXEC', 
 'PAVIMENTAÇÃO - FINANCEIRO(R$)',
 'DUPLICAÇÃO - Descrição',
 'DUPLICAÇÃO - km(i)', 
 'DUPLICAÇÃO - km(f)', 
 'DUPLICAÇÃO - Ext.(km)', 
 'DUPLICAÇÃO - PERCENTUAL (%) PLAN', 
 'DUPLICAÇÃO - (km)% PLAN', 
 'DUPLICAÇÃO - FINANCEIRO PLAN (R$)', 
 'DUPLICAÇÃO - PERCENTUAL (%) EXEC', 
 'DUPLICAÇÃO - (km)% EXEC', 
 'DUPLICAÇÃO - FINANCEIRO(R$)',
 'OAE - Descrição',
 'OAE - km(i)', 
 'OAE - km(f)', 
 'OAE - Ext.(km)', 
 'OAE - PERCENTUAL (%) PLAN', 
 'OAE - (km)% PLAN', 
 'OAE - FINANCEIRO PLAN (R$)', 
 'OAE - PERCENTUAL (%) EXEC', 
 'OAE - (km)% EXEC', 
 'OAE - FINANCEIRO(R$)',
 'CONTORNO - Descrição',
 'CONTORNO - km(i)', 
 'CONTORNO - km(f)', 
 'CONTORNO - Ext.(km)', 
 'CONTORNO - PERCENTUAL (%) PLAN', 
 'CONTORNO - (km)% PLAN', 
 'CONTORNO - FINANCEIRO PLAN (R$)', 
 'CONTORNO - PERCENTUAL (%) EXEC', 
 'CONTORNO - (km)% EXEC',
 'CONTORNO - FINANCEIRO(R$)',
 'FX ADICIONAL - Descrição',
 'FX ADICIONAL - km(i)',
 'FX ADICIONAL - km(f)', 
 'FX ADICIONAL - Ext.(km)', 
 'FX ADICIONAL - PERCENTUAL (%) PLAN', 
 'FX ADICIONAL - (km)% PLAN', 
 'FX ADICIONAL - FINANCEIRO PLAN (R$)', 
 'FX ADICIONAL - PERCENTUAL (%) EXEC', 
 'FX ADICIONAL - (km)% EXEC', 
 'FX ADICIONAL - FINANCEIRO(R$)',
 'TERCEIRA FAIXA - Descrição',
 'TERCEIRA FAIXA - km(i)', 
 'TERCEIRA FAIXA - km(f)', 
 'TERCEIRA FAIXA - Ext.(km)', 
 'TERCEIRA FAIXA - PERCENTUAL (%) PLAN', 
 'TERCEIRA FAIXA - (km)% PLAN', 
 'TERCEIRA FAIXA - FINANCEIRO PLAN (R$)', 
 'TERCEIRA FAIXA - PERCENTUAL (%) EXEC', 
 'TERCEIRA FAIXA - (km)% EXEC', 
 'TERCEIRA FAIXA - FINANCEIRO(R$)']
```

```python
# Cria um dicionário que mapeia qualquer coluna a partir do índice 9 para os novos nomes
rename_dict = {df_concatenado.columns[i]: name for i, name in enumerate(new_column_names, start=9)}

# Usa o método rename para renomear as colunas, substituindo os nomes existentes
df_concatenado.rename(columns=rename_dict, inplace=True)
```

```python
# Lista de strings que devem estar no nome das colunas para serem mantidas
keywords_to_keep = [
    'ID-ÚNICO', 'SETOR', 'EMPREENDIMENTO', 
    'Descrição', 'Ext.(km)', 'FINANCEIRO PLAN (R$)', 'FINANCEIRO(R$)', 
    '(km)% PLAN', '(km)% EXEC'
]

# Identifica colunas que possuem qualquer uma das strings na lista de keywords
columns_to_keep = [col for col in df_concatenado.columns if any(keyword in col for keyword in keywords_to_keep)]
```

```python

# Mantém apenas as colunas identificadas
df_concatenado = df_concatenado[columns_to_keep]
```

```python
df_concatenado.columns
```

```python
df_concatenado
```

```python
df_concatenado1 = df_concatenado[['ID-ÚNICO', 'SETOR', 'EMPREENDIMENTO']]
```

```python
colunas_manter = ['ID-ÚNICO', 'SETOR', 'EMPREENDIMENTO']

# Use 'difference' para obter as colunas a serem unpivotadas
colunas_unpivot = df_concatenado.columns.difference(colunas_manter)

# Realizar o unpivot (melt) das outras colunas
df_concatenado = df_concatenado.melt(id_vars=colunas_manter, value_vars=colunas_unpivot,
                    var_name='Atributo', value_name='Valor')
```

```python
df_concatenado['SETOR'] = 'Rodoviário'
```

```python
df_concatenado[['Atributo.1', 'Atributo.2']] = df_concatenado['Atributo'].str.split(' - ', expand=True)
```

```python
df_concatenado.columns
```

```python
df_concatenado = df_concatenado[['ID-ÚNICO', 'SETOR', 'EMPREENDIMENTO',
       'Atributo.1', 'Atributo.2', 'Valor']]
```

```python
df_concatenado.dropna(subset=['Valor'])
df_concatenado = df_concatenado.loc[df_concatenado['Valor'] != 0]
```

```python
# Pivotar o DataFrame
df_concatenado = df_concatenado.pivot_table(index=['ID-ÚNICO', 'Atributo.1'], 
                                            columns='Atributo.2', values='Valor', aggfunc='first').reset_index()
```

```python
df_concatenado.columns
```

```python
df_concatenado = df_concatenado.drop(columns='ID-ÚNICO')
```

```python
# Concatenar os DataFrames
df_concatenado = pd.concat([df_concatenado1, df_concatenado], axis=1)
```

```python
df_concatenado = df_concatenado.dropna(subset=['Atributo.1'])
```

```python
df_concatenado
```

```python
df_concatenado.columns
```

```python
df_concatenado = df_concatenado[['ID-ÚNICO', 'SETOR', 'EMPREENDIMENTO',
       'Atributo.1', 'Descrição', 'Ext.(km)', 'FINANCEIRO PLAN (R$)', 'FINANCEIRO(R$)', 
    '(km)% PLAN', '(km)% EXEC']]
```

```python
# Renomear Colunas
df_concatenado.rename(columns={'ID-ÚNICO': 'ID-ÚNICO2'}, inplace=True)
df_concatenado.rename(columns={'SETOR': 'SETOR2'}, inplace=True)
df_concatenado.rename(columns={'EMPREENDIMENTO': 'EMPREENDIMENTO2'}, inplace=True)
df_concatenado.rename(columns={'Descrição': 'Descrição2'}, inplace=True)
df_concatenado.rename(columns={'Ext.(km)': 'Ext.(km)2'}, inplace=True)
df_concatenado.rename(columns={'FINANCEIRO PLAN (R$)': 'FINANCEIRO PLAN (R$)2'}, inplace=True)
df_concatenado.rename(columns={'FINANCEIRO(R$)': 'FINANCEIRO(R$)2'}, inplace=True)
df_concatenado.rename(columns={'(km)% PLAN': '(km)% PLAN2'}, inplace=True)
df_concatenado.rename(columns={'(km)% EXEC': '(km)% EXEC2'}, inplace=True)
```

```python
df_concatenado
```

```python
file_path_plan_original = 'NEW-SCRIPT - LIMPO\PLAN_EXEC ATÉ ANO ANTERIOR (22).xlsx'
df_plan_original = pd.read_excel(file_path_plan_original)
```

```python
df_plan_original
```

```python
df_final = pd.concat([df_plan_original, df_concatenado], ignore_index=True)
```

```python
df_final
```

```python
df_final.to_excel('NEW-SCRIPT - LIMPO/PLAN_EXEC ATÉ ANO ANTERIOR (22).xlsx', index=False)
```
